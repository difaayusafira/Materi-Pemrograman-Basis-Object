```python
#contoh polymorphisme

class Shark():
    def swim(self):
        print ("The Shark is swimming")
    def swim_backwards(self):
        print ("the Shark cannot swim backwars, can shink backwards")
    def skeleton(self):
        print ("The Shark's skeleton is made of cartilage")
class Clownfish():
    def swim(self):
        print ("The Clownfish is swimming")
    def swim_backwards(self):
        print ("The Clownfish can swim backwards")
    def skeleton(self):
        print ("The Clownfish's skeleton is made of bone")

sammy = Shark()
sammy.skeleton()

casey = Clownfish()
casey.skeleton()

#Polymorphsim wih function
def in_the_pasific(fish):
    fish.swim()
sammy = Shark()
casey = Clownfish()
in_the_pasific(sammy)
in_the_pasific(casey)

#Polymorphsim with class method
sammy = Shark()
casey = Clownfish()
for fish in(sammy,casey):
    fish.swim()
    fish.swim_backwards()
    fish.skeleton()
```

```python
class Bear():
    def sound(self):
        print ("Groooaaarrrrr")
class Dog():
    def sound(self):
        print ("wooff wooff")
def makeSound(animalType):
    animalType.sound()
bearObj = Bear()
dogObj = Dog()

makeSound(bearObj)
makeSound(dogObj)
```
