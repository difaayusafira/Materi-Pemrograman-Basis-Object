~ Inheritance adalah pewarisan/penurunan
~ Inheritance adalah konsep pemrograman dimana sebuah class dapat 
                   menurunkan property dan method yang dimilikinya 
                   kepada class lain.
Class yang akan diturunkan disebut sebagai class induk (parent class) 
Class yang menerima penurunan disebut sebagai class anak (child class)

Inheritance memiliki 4 tipe :
    1. Default = property,property, dan method
    diwariskan semua tanpa dirubah dan tanpa penambahan, diberikan dari parent   
    ke anak. 
    2. Overriding(harus menyamakan atribut dan anak) adalah sifat dari anak 
    digantikan/dimodifikasikan property dan method. 
    yang dioveride adalah property/anak kelas atau isi dari method tersebut
    3. Super adalah menambahkan property yang diletakkan di child class dan 
    digunakan di constructor.
    Inisialisasi dikelas anak harus sesuai dengan berapa banya parameter di 
    constructor kelas anak.
    4. Multiple, digunakan ketika suatu kelas dapat mewarisi atribut dan method
    property yang ada di induk diwariskan kepada anak.
    Memungkinkan menggunakan super class, tetapi tidak menggunakan overriding, 
    karena multiple ini menggunakan minimal dua parent dan biasanya tidak sama. 
    dan yang diwariskan adalah method 


```python
#contoh default

class Fish: 
    def __init__(self, first_name, last_name="Fish", skeleton="bone", eyelids=False): 
        self.first_name = first_name 
        self.last_name = last_name 
        self.skeleton = skeleton 
        self.eyelids = eyelids 
    def swim(self): 
        print("The fish is swimming.") 
    def swim_backwards(self): 
        print("The fish can swim backwards.") 
class Trout(Fish):
    pass
class Clownfish(Fish): 
    def live_with_anemone(self): 
        print("The clownfish is coexisting with sea anemone.") 
```

```python
#contoh overriding

class persegi:
    def __init__(self,segi,bentuk = "Persegi"): #Diberi pendefisian
        self.segi = segi
        self.bentuk = bentuk
    def display(self):
        print ("Jumlah Segi = ", self.segi,"Bentuk = ", self.bentuk)
    def deskripsi(self):
        print ("4 sisi sama panjang ")
class persegipanjang(persegi):
    def __init__(self,segi,bentuk = "Persegi Panjang"):
        self.segi = segi
        self.bentuk = bentuk
    def deskripsi (self):
        print ("Ada 2 pasang yang panjang sisi sama")
class bujursangkar(persegi):
    def __init__(self,segi,bentuk = "Bujur Sangkar"):
        self.segi = segi
        self.bentuk = bentuk
    def deskripsi(self):
        print ("Keduanya sama")
    

a = persegi(4)
a.display()
a.deskripsi()
b = persegipanjang(4)
b.display()
b.deskripsi()
c = bujursangkar(4)
c.display()
c.deskripsi()
```

```python
#contoh super class

class user: #parent
    def __init__(self,name): #constructor adalah inisialisasi dari properti
        self.name = name
    def printName(self):
        print ("Name = " + self.name)
class Programmer(user): #baca di Inheritance SuperClass
    def __init__(self,name,lastname):
        user.__init__(self,name)
        self.lastname = lastname
    def printlastname(self):
        print ("Name = " + self.lastname)     
    def doPython(self):
        print("Programming Python")
class Programmer1(user):
    pass #semua yang diclass bisa digunakan 

brian = user("brian")
brian.printName()

diana1 = Programmer1("diana")
diana1.printName()

diana = Programmer("diana")
diana.printName()
diana.doPython()
```

```python
#contoh multiple

class person:
    def __init__(self):
        self.name = input("Name = ")
    
    def display(self):
        print("Name = ", self.name)
    
class marks:
    def __init__(self):
        self.npm = input("NPM = ")
        print ("Nilai")
        self.math = int(input("Math = "))
        self.biologi = int(input("Biologi = "))
    def display(self):
        print("NPM = ", self.npm)
        print("Nilai = ", self.math + self.biologi)
class student(person,marks):
    def __init__(self):
        person.__init__(self)
        marks.__init__(self)
    def result(self):
        person.display(self)
        marks.display(self)
stu1 = student()
stu2 = student()
stu1.result()
stu2.result()
```
